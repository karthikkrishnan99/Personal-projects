```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
plt.style.use('seaborn-whitegrid')
import numpy as np
from sklearn import preprocessing
```

```python
importdf = pd.read_excel('India China Import (1).xlsx')
exportdf = pd.read_excel('India - China exports (4).xlsx')
```

```python
exportdf.drop([21,22],axis=0,inplace=True)
```

```python
importdf
```

```python
exportdf
```

```python
listyrs=[]
for i in range(5,importdf.shape[1]):
    listyrs.append(i)
importlist=importdf.iloc[0,listyrs].tolist()
```

```python
listyrs=[]
for i in range(5,exportdf.shape[1]):
    listyrs.append(i)
exportlist=exportdf.iloc[0,listyrs].tolist()
```

```python
for i in range(0,len(importlist)):
    importlist[i]=importlist[i]/(1000000)
    exportlist[i]=exportlist[i]/(1000000)
```

```python
difflist_ime=[] #ime-import minus export , emi- export minus import, ebi-export by import
difflist_emi=[]
difflist_ebi=[]
for i in range(0,len(importlist)):
    difflist_ime.append(importlist[i]-exportlist[i])
    difflist_emi.append(exportlist[i]-importlist[i])
    difflist_ebi.append(exportlist[i]/importlist[i])
```

```python
fig = plt.figure()
ax = plt.axes()
list1=[]
for i in range(2008,2019):
    list1.append(i)
plt.plot(list1,importlist,color='r')
plt.plot(list1,exportlist,color='g')
plt.plot(list1,difflist_emi,color='b')
plt.axhline(linewidth=2, color='y')
red_patch = mpatches.Patch(color='r', label='Import in billion $')
green_patch=mpatches.Patch(color='g', label='Export in billion $')
blue_patch=mpatches.Patch(color='b', label='Net Trade in billion $')
plt.legend(handles=[red_patch,green_patch,blue_patch])
plt.title('India China Trade Statistics')
plt.show()
```

```python
importcopy=importdf.copy()
exportcopy=exportdf.copy()
```

```python
for i in range(5,importcopy.shape[1]):
    colval=importcopy.iloc[0,i]
    for j in range(0,importcopy.shape[0]):
        colname=importcopy.columns[i]
        importcopy.at[j,colname]=(importcopy.iloc[j,i]/colval)
```

```python
for i in range(5,exportcopy.shape[1]):
    colval=exportcopy.iloc[0,i]
    for j in range(0,exportcopy.shape[0]):
        colname=exportcopy.columns[i]
        exportcopy.at[j,colname]=(exportcopy.iloc[j,i]/colval)
```

```python
importcopy.to_csv('normalised import.csv',index=False)
exportcopy.to_csv('normalised export.csv',index=False)
```

```python
fig = plt.figure()
ax = plt.axes()
list1=[]
for i in range(2008,2019):
    list1.append(i)
plt.plot(list1, difflist_ebi, color='r')
red_patch = mpatches.Patch(color='r', label='Export to Import Ratio')
plt.legend(handles=[red_patch])
plt.title('India China Export to Import Ratio')
plt.show()
```
